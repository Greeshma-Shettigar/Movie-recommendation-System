```python
import numpy as np
import pandas as pd
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
movies.head(1)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies=movies.dropna()
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
           L.append(i['name'])
           counter+=1
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
           L.append(i['name'])
           break
    return L
# from crew we are only fetching the director name
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#overview is in string format we will convert that into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#remove the space between the  words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#create new column called tags in movies by concatanating overview,genres,keywords,cast,crew
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
#convert the tag list into string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#convert tags into lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)          #convert into string again
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```
